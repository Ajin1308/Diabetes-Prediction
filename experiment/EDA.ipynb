```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
```

```python
rain_df = pd.read_csv("weatherAUS.csv")
rain_df.head()
```

```python
plt.figure(figsize = (20,30))
sns.set(style ='whitegrid')

sns.countplot(x ='RainTomorrow', hue='Location', data = rain_df)
plt.title("Distribution of RainTomorrow by Location")
plt.xlabel("Rain Tomorrow")
plt.ylabel("Count")
plt.show()
```

```python
fig = plt.figure(figsize=(20,8),facecolor='white')

gs = fig.add_gridspec(1,2)

ax = [_ for _ in range(2)]
ax[0] = fig.add_subplot(gs[0,0])
ax[1] = fig.add_subplot(gs[0,1])

ax[0].text(-1,15,"Relationship between Rainfall and Sunshine", fontsize = 17,
           fontweight='normal', fontfamily='Verdana')

ax[1].text(-1,430,"Relationship between Rainfall and WindGustSpeed",fontsize = 17,
           fontweight='normal', fontfamily='Verdana')

sns.lineplot(data = rain_df, x ='Sunshine', y = 'Rainfall', ax = ax[0])
sns.scatterplot(data = rain_df, x = 'WindGustSpeed', y = 'Rainfall', ax=ax[1])

for i  in range(2):
    ax[i].set_ylabel('Rainfall').set_rotation(0)
    ax[i].set_yticklabels('')
    #Spines -> border
    for direction in ['top','right','left']:
        ax[i].spines[direction].set_visible(False)
        
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(20,8),facecolor='white')

gs = fig.add_gridspec(1,2)

ax = [_ for _ in range(2)]
ax[0] = fig.add_subplot(gs[0,0])
ax[1] = fig.add_subplot(gs[0,1])

ax[0].text(-1,15,"Relationship between Sunshine and Evaporation", fontsize = 17,fontweight='normal', fontfamily='Verdana')

ax[1].text(-1,430,"Relationship between Cloud3mp and Rainfall",fontsize = 17,fontweight='normal', fontfamily='Verdana')

sns.lineplot(data = rain_df, x ='Sunshine', y = 'Evaporation', ax = ax[0], hue = 'RainTomorrow')
sns.scatterplot(data=rain_df,x='Cloud3pm', y='Rainfall',hue='RainTomorrow',ax=ax[1])

for i in range(2):
    ax[i].set_yticklabels([])
    #Spines -> border
    for direction in ['top','right','left']:
        ax[i].spines[direction].set_visible(False)
        
```
